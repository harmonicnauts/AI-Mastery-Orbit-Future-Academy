# 3. Preparing the data

```python
image_datagen = tf.keras.preprocessing.image.ImageDataGenerator(
    rescale=1.0/255.0,
    validation_split=0.2,
    shear_range=0.1,
    zoom_range=0.2,
    horizontal_flip=True,
    vertical_flip=True,
    width_shift_range=0.2,
    height_shift_range=0.2,
    brightness_range=(0.8,1.1)
)
```

```python
train_ds = image_datagen.flow_from_directory(
        path+'/train',
        subset='training',
        target_size=(640 , 640),
        batch_size=16)

val_ds = image_datagen.flow_from_directory(
        path+'/train',
        subset='validation',
        target_size=(640 , 640),
        batch_size=16)
```

# 4. Building the model

```python
model_Xception = tf.keras.applications.Xception(input_shape=(640 , 640, 3),
                                                 include_top=False,
                                                 weights='imagenet')
model_Xception.trainable=False
```

```python
model = tf.keras.Sequential()

model.add(model_Xception)
model.add(BatchNormalization())

model.add(Conv2D(128, (3,3), activation='relu', kernel_regularizer=tf.keras.regularizers.L2(0.001)))
model.add(Dropout(0.5))

model.add(GlobalAveragePooling2D())
model.add(Dropout(0.5))

model.add(Dense(4096, activation='relu'))
model.add(Dropout(0.5))
model.add(Dense(2048, activation='relu', kernel_regularizer=tf.keras.regularizers.L2(0.001)))
model.add(Dropout(0.5))
model.add(Dense(1024, activation='relu', kernel_regularizer=tf.keras.regularizers.L2(0.001)))
model.add(Dropout(0.5))
model.add(Dense(4, activation='softmax'))

model.summary()
```

```python
# checkpoint_filepath = '/content/drive/MyDrive/Kuliah/MSIB/Computer_Vision/Dataset/ofa-ai-mastery-computer-vision/Checkpoint/darknet53.conv.74'
# model_checkpoint_callback = tf.keras.callbacks.ModelCheckpoint(
#     filepath=checkpoint_filepath,
#     save_weights_only=True,
#     monitor='val_accuracy',
#     mode='max',
#     save_best_only=True)

model.compile(optimizer=tf.keras.optimizers.Adam(learning_rate=0.00005), loss='categorical_crossentropy', metrics=['accuracy'])

model.summary()
```

```python
tf.keras.utils.plot_model(model, show_shapes=True)
```

```python
save_best = tf.keras.callbacks.ModelCheckpoint("Model.h5",monitor='val_accuracy',save_best_only=True, verbose=1)
```

# 5. Training and evaluating

```python
history_Xcept = model.fit(train_ds, validation_data=val_ds, epochs=8, callbacks=[save_best], shuffle=True)
```

```python
model.evaluate(val_ds)

plt.figure()
plt.plot(history_Xcept.history['loss'])
plt.plot(history_Xcept.history['val_loss'])
plt.legend(['loss','val_loss'],loc='upper right')
plt.show()
plt.figure()
plt.plot(history_Xcept.history['accuracy'])
plt.plot(history_Xcept.history['val_accuracy'])
plt.legend(['accuracy','val_accuracy'],loc='upper right')
plt.show()
```

Dapat dilihat bahwa tidak ada tanda bahwa model overfit / underfit. Hasil akurasi model mencapai nilai akhir yaitu sekitar 98-99%

# 6. Preparing for submission

```python
sub_csv ='./Sample_submission.csv'
path_test = './ofa-ai-mastery-computer-vision/test/test'
```

```python
df_sub = pd.read_csv(sub_csv)
image_id = df_sub['ID']
df_sub.head(10)
```

```python
# each image will be treated individually and saved in a list which is then merged with the id of it's correspondence

from tensorflow.keras.preprocessing.image import load_img


# making a list for our predictions
labels = []
count = 0

for i in image_id:
    print(count+1)
    image = load_img(path_test+'/'+str(i)+'.jpg', target_size=(640,640))
    img = np.array(image)
    img = img / 255.0
    img = img.reshape(1,640,640,3)
    label = model.predict(img)
    label_id = label[0].tolist()
    
    labels.append(label_id.index(max(label_id)))
    count+=1
print(count)
```

```python
# we will replace the previous empty columns with the new predictions    
    
df_sub['Label'] = labels
df_sub.head(10)
```

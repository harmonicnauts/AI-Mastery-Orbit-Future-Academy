## Mengimpor Library yang diperlukan

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

# Memuat Dataset

```python
dataset = pd.read_csv('../Dataset/Wine.csv')
```

```python
dataset
```

# Membuat variabel dependen dan variabel independen

```python
X = dataset.iloc[:,:13].values # Menyimpan kolom 1 - 12 di var X sebagai var independen
y = dataset.iloc[:, 13].values# Menyimpan kolom 13 di var y sebagai var independen
```

```python
X
```

```python
y
```

# Split data menjadi train dan test

```python
from sklearn.model_selection import train_test_split
# Membagi data menjadi 4 var berikut dengan proporsi data test nya adalah 30% dari total data
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size = 0.3, random_state = 0)
```

```python
X_train
```

```python
X_test
```

```python
y_train
```

# Proses Feature Scaling

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
```

```python
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

```python
X_test
```

# Run algoritma PCA

```python
from sklearn.decomposition import PCA
pca = PCA(n_components = None)
X_train = pca.fit_transform(X_train)
X_test = pca.transform(X_test)
hasil_variance = pca.explained_variance_ratio_ # Mengecek variance
```

```python
print(hasil_variance, f'sum = {sum(hasil_variance)}')
```

# Proses Pemilihan PC

```python
var_percent = np.round(pca.explained_variance_ratio_ * 100, decimals = 1)

labels = ['PC' + str(char) for char in range(1, len(var_percent) + 1)]
plt.bar(x=range(1, len(var_percent) + 1), height = var_percent, tick_label = labels)
plt.xlabel('Principal Component (PC)')
plt.ylabel('Persentase Variance')
plt.title('Screeplot')
```

```python
print(f'PC1 + PC2 = {pca.explained_variance_ratio_[0] * 100 + pca.explained_variance_ratio_[1] * 100}%')
```

Karena PC1 + PC2 sudah mewakili sekitar 56% dari data maka bisa digunakan untuk membuat model tanpa PC3 sampai setelahnya

```python
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size = 0.3, random_state = 0)
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
pca = PCA(n_components = 2)
X_train = pca.fit_transform(X_train)
X_test = pca.transform(X_test)
hasil_variance_2 = pca.explained_variance_ratio_
```

```python
hasil_variance_2
```

```python
X_train
```

```python
y_train
```

# Run algoritma Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
logs_clf = LogisticRegression(random_state=0)
logs_clf.fit(X_train, y_train)
```

```python
y_pred = logs_clf.predict(X_test)
```

```python
y_pred
```

# Confusion Matrix

```python
y_test.shape, y_pred.shape
```

```python
from sklearn.metrics import confusion_matrix
import scikitplot.metrics as skplt
skplt.plot_confusion_matrix(y_test, y_pred, figsize = (5,5))
```

```python
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test, y_pred)
cm
```

# Visualisasi Training Set

```python
# Visualisasi Training Set
from matplotlib.colors import ListedColormap
X_set, y_set = X_train, y_train
X1, X2 = np.meshgrid(np.arange(start = X_set[:, 0].min() - 1, stop = X_set[:, 0].max() + 1, step = 0.01),
                     np.arange(start = X_set[:, 1].min() - 1, stop = X_set[:, 1].max() + 1, step = 0.01))
plt.contourf(X1, X2, logs_clf.predict(np.array([X1.ravel(), X2.ravel()]).T).reshape(X1.shape),
             alpha = 0.75, cmap = ListedColormap(('red', 'green', 'blue')))
plt.xlim(X1.min(), X1.max())
plt.ylim(X2.min(), X2.max())
for i, j in enumerate(np.unique(y_set)):
    plt.scatter(X_set[y_set == j, 0], X_set[y_set == j, 1],
                c = ListedColormap(('red', 'green', 'blue'))(i), label = j)
plt.title('Logistic Regression (Training set)')
plt.xlabel('PC1')
plt.ylabel('PC2')
plt.legend()
plt.show()
```

```python
# Visualisasi Test Set
from matplotlib.colors import ListedColormap
X_set, y_set = X_test, y_test
X1, X2 = np.meshgrid(np.arange(start = X_set[:, 0].min() - 1, stop = X_set[:, 0].max() + 1, step = 0.01),
                     np.arange(start = X_set[:, 1].min() - 1, stop = X_set[:, 1].max() + 1, step = 0.01))
plt.contourf(X1, X2, logs_clf.predict(np.array([X1.ravel(), X2.ravel()]).T).reshape(X1.shape),
             alpha = 0.75, cmap = ListedColormap(('red', 'green', 'blue')))
plt.xlim(X1.min(), X1.max())
plt.ylim(X2.min(), X2.max())
for i, j in enumerate(np.unique(y_set)):
    plt.scatter(X_set[y_set == j, 0], X_set[y_set == j, 1],
                c = ListedColormap(('red', 'green', 'blue'))(i), label = j)
plt.title('Logistic Regression (Test set)')
plt.xlabel('PC1')
plt.ylabel('PC2')
plt.legend()
plt.show()
```
